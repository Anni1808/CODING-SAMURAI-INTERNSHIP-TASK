```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#aesthetic parameters
sns.set(style="whitegrid")
%matplotlib inline
```

```python
#load dataset
titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
print("Shape: ", titanic.shape)
```

```python
titanic.info()
```

```python
#summary statistics
titanic.describe(include='all')
```

```python
titanic.isnull().sum()
```

```python
#visualize missing data
plt.figure(figsize=(10, 6)) #width=10inches, height=6inches
sns.heatmap(titanic.isnull(), cbar=True, cmap="Blues") #returns a df of same shape
#cbar = color bar legend
plt.title("Missing values in Titanic Dataset")
plt.show()
```

```python
#countplot for categorical values
sns.countplot(x='sex', data=titanic)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='class', data=titanic)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='survived', data=titanic)
plt.title("Survived Count")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
#histogram for age
sns.histplot(titanic['age'].dropna(), bins=30, kde=True) #bins = number of lines on graph
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=titanic['fare'])
plt.title("Fare Distribution")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Count by Gender")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival Count by Class")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.histplot(data=titanic, x='age', hue='survived', bins=30, kde=True, element="step")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
numeric_data = titanic.select_dtypes(include=[np.number])

corr_matrix = numeric_data.corr()
#It shows the correlation coefficients between all pairs of numerical columns

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
#cmap = chooses how colors represent correlation.
#linewidth = Adds thin lines between cells, improving visual separation and readability.
plt.title("Correlation Matrix")
plt.show()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

```python
titanic.dropna(subset=['embarked'], inplace=True)
```
